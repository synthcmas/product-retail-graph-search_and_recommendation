```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
##loading the uselful chunk of transactions
transactions_i=pd.read_csv('transactions_train.csv',usecols=['customer_id','article_id'])
transactions_i.head()
```

```python
transactions_i.shape
```

```python
##loading the useful chunk from customers
customers_i=pd.read_csv('customers.csv',usecols=['customer_id','postal_code'])
customers_i.head()
```

```python
##loading the useful chunk from articles
articles_i=pd.read_csv('articles.csv',usecols=['article_id','product_group_name','index_group_name'])
articles_i.head()
```

```python
##merging transactions_i and customers_i
one=pd.merge(transactions_i,customers_i,on='customer_id')
one.head()
```

```python
##merging one and articles_i
two=pd.merge(one,articles_i,on='article_id')
del(transactions_i,customers_i,articles_i,one)
two.head()
```

```python
##stratified split of the data
from sklearn.model_selection import StratifiedShuffleSplit
X=two.drop(['product_group_name','index_group_name'],axis=1)
y=two[['product_group_name','index_group_name']]
del(two)
sss=StratifiedShuffleSplit(test_size=0.95,train_size=0.05,random_state=42)
```

```python
train_test=next(sss.split(X,y))
```

```python
train_idx=train_test[0]
```

```python
transactions_f=pd.read_csv('transactions_train.csv')
transactions=transactions_f.loc[train_idx]
del(X,y,transactions_f)
transactions.reset_index(inplace=True)
transactions.drop('index',axis=1,inplace=True)
transactions.head()
```

```python
##importing customers and articles data
customers=pd.read_csv('customers.csv',usecols=['customer_id','club_member_status','fashion_news_frequency','age','postal_code'])
articles=pd.read_csv('articles.csv')
##merging transactions and customers data
first=pd.merge(transactions,customers,on='customer_id')
##deleting transactions and customers
del(transactions,customers)
##merging all
final=pd.merge(first,articles,on='article_id')
final.head()
```

```python
final.shape
```

```python
final.to_csv('stratified_sample_1.csv',index_label=False)
```
