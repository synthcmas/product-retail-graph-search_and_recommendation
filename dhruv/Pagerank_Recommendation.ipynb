```python
# making a graph instance and connecting to database
graph = Graph()
remoteConn = DriverRemoteConnection('wss://database-prg-instance-1.cbtcet4rvkih.eu-central-1.neptune.amazonaws.com:8182/gremlin','g')
g = graph.traversal().withRemote(remoteConn)
```

```python
root = 'Sim mega hood'
```

```python
def matching_node_value(root):
    x_value = g.V().has('value',root).both().values().toList()
    x_label = g.V().has('value',root).both().label().toList()
    return x_label, x_value
    
```

```python
def all_common_product( node_neighbourbood,x_label):
    neighbour_product = []
    for i in x_label:
        neighbour_product = neighbour_product + node_neighbourbood[i] 
    neighbour_product = set(neighbour_product)
    return neighbour_product
        
```

```python
def neighbour_attr_node_match(neighbour_product, node_neighbourbood, x_label, x_value):
    
    prod_dict = {}
    prod_score = {}
    for i in neighbour_product:
        attributes = []
        for label in x_label:
            if i in node_neighbourbood[label]: attributes.append(label)
            
        prod_dict[i] = attributes
        prod_score[i] = len(attributes)
    
    return prod_dict, prod_score
    
    
```

```python
def pagerank_recommendation(root):
    node_neighbourbood = {}
    x_label, x_value = matching_node_value(root)
    for x,y in zip(x_label, x_value):
        product_temp = []
        print(x,y)
        if(x == 'subcategory'):
            product = g.V().hasLabel(x).has('value',y).out('sub_of').path().toList()
            for i in product:product_temp.append( g.V().has('~id' , i[1].id).path().by('value').toList()[0][0])
            node_neighbourbood[x] = product_temp
            print("Done \n")
        
        if(x != 'subcategory'):
            product = g.V().hasLabel(x).has('value',y).both().path().toList()
            for i in product:product_temp.append( g.V().has('~id' , i[1].id).path().by('value').toList()[0][0])
            node_neighbourbood[x] = product_temp
            print("Done \n")
    
        
    return x_label, x_value, node_neighbourbood

            
 
        
    
    
```

```python
x_label, x_value, node_neighbourbood =  pagerank_recommendation(root)
```

```python
neighbour_product = all_common_product( node_neighbourbood,x_label)
```

```python
prod_dict, prod_score = neighbour_attr_node_match(neighbour_product, node_neighbourbood, x_label, x_value)
```

```python
prod_score_sorted = sorted( prod_score.items(), key=lambda kv:
                 (kv[1], kv[0]),reverse=True)
```
